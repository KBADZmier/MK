```python
import numpy as np
import matplotlib.pyplot as plt

def elu(z, alpha=1.0):
    return np.where(z >= 0, z, alpha * (np.exp(z) - 1))

def elu_gradient(z, alpha=1.0):
    return np.where(z >= 0, 1, alpha * np.exp(z))

# Zakres danych
z = np.linspace(-5, 5, 500)
y = elu(z)
dy = elu_gradient(z)

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
plt.plot(z, y, label='Funkcja ELU', color='blue', linewidth=2)
plt.plot(z, dy, label='Gradient ELU', color='red', linestyle='--', linewidth=2)

# Konfiguracja wykresu
plt.title('Funkcja aktywacji ELU i jej gradient')
plt.xlabel('z')
plt.ylabel('Warto')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()
```

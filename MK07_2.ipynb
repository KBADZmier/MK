```python
import numpy as np

# Funkcje aktywacji i ich pochodne
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1.0 - np.tanh(x)**2

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Inicjalizacja wag
np.random.seed(42)
W1 = np.random.randn(2, 2)
b1 = np.zeros((1, 2))
W2 = np.random.randn(2, 1)
b2 = np.zeros((1, 1))


X = np.array([[0.5, -0.2]])
y_true = np.array([[1.0]])


Z1 = np.dot(X, W1) + b1
A1 = tanh(Z1)
Z2 = np.dot(A1, W2) + b2
A2 = relu(Z2)

# Funkcja straty (MSE)
loss = np.mean(np.square(A2 - y_true))

# --- BACKWARD PROPAGATION (Obliczanie gradient√≥w) ---

dA2 = 2 * (A2 - y_true)

# Gradienty dla Warstwy 2 (ReLU)
dZ2 = dA2 * relu_derivative(Z2)
dW2 = np.dot(A1.T, dZ2)
db2 = np.sum(dZ2, axis=0, keepdims=True)

# Gradienty dla Warstwy 1 (tanh)
dA1 = np.dot(dZ2, W2.T)
dZ1 = dA1 * tanh_derivative(Z1)
dW1 = np.dot(X.T, dZ1)
db1 = np.sum(dZ1, axis=0, keepdims=True)

print("Gradienty dla Warstwy 2:")
print("dW2:", dW2)
print("db2:", db2)
print("\nGradienty dla Warstwy 1:")
print("dW1:", dW1)
print("db1:", db1)
```
